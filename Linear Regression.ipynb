#### What is Linear Regression

* A linear regression is to predict a value of dependent variable given the independent variable(s)
* Example:
    * y = Ax + B
        * Coefficient/slope: A
        * Bias/intercept: B
    * x is independent variable: for example weight
    * y is dependent variable: for example height

### Example of Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

# Generate a random values of x and y (y values depends on the x values)
## Example: X represents distance travelled by a car(s)
## Example: Y represents cost of car(s) dependent on the travelled distance
n = 40 # number of cars
x = np.random.randn(n) # distance of the car(s) 
y = x * np.random.randn(n) # cost of the car(s)

# plot these values
colors = np.random.rand(n)
## np.poly1d => turns the array into polyominal equation
## np.polyfit => minimizes the squared error to get the "best-fitting" line
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))

plt.scatter(x, y, c=colors, alpha=0.5)
plt.xlabel('x axes (distance travelled)')
plt.ylabel('y axes (cost)')
plt.show()
```

### Goal of Linear Regression

* Generate a "Line of best-fit" or "Best-fitting Line" or "Regression line"
    * The line in figure above represents "Regression Line"
    * This line "minimizes the "sum of squared error" (**Big word alert**) of prediction" (**Confusing sentence alert**)
        * "Sum of Squared Error" => (prediction - actual value)^2
        * Basically, this line predicts the y values (in our case, the cost of the car(s) for the future)
        * By using this concept, you or others can make a decision about buying a car(s) in the future or use same concept for other predictions(but seriously try to get real data not some made by data by me :D)
